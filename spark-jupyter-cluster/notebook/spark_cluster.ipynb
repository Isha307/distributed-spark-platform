```python
from pyspark.sql import SparkSession

spark = (
    SparkSession.builder
    .appName("Test Spark")
    .master("spark://spark-master:7077")
    .getOrCreate()
)
```

```python
import os
os.makedirs("/data/processed", exist_ok=True)
```

```python
import os
os.listdir("/data")
```

```python
os.listdir("/input")
```

```python
df = (
    spark.read
    .option("header", True)
    .csv("/input/raw/employees.csv")
)

df.show()
```

```python
df.printSchema()
```

```python
df.rdd.getNumPartitions()
```

```python
from pyspark.sql.functions import col
 
df = df.withColumn("bonus", col("salary")*1.1)
```

```python
df.show()
```

```python
df.write.mode("overwrite").parquet("/data/processed/employees_parquet")
```
